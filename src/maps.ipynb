```python
# 1. Loop the dataframe
    #if the day: holiday
        #icon diferent
    #marker
    #marker.add_to(map)
    
df.sample()

#How to loop a df?
for index, row in df.iterrows():
    
    #1. location of the accident
    lat = row["lat"]
    lon = row["lon"]
    
    #dict_coordinates = {"location": [row["lat"], row["lon"]]}
        #Marker(location = [])
    
    if row["Type"] == "Design Companies":
        icon = Icon (
            icon = "sitemap",
            prefix = "fa",
            color = "blue",
            icon_color = "white"
        )
        
    elif row["Type"] == "Starbucks":
        icon = Icon (
                icon = "coffee",
                prefix = "fa",
                color = "green",
                icon_color = "white"
        )

    elif row["Type"] == "Highschool":
        icon = Icon (
                icon = "briefcase",
                prefix = "fa",
                color = "purple",
                icon_color = "white"
        ) 
            
    elif row["Type"] == "Middle school":
        icon = Icon (
                icon = "television",
                prefix = "fa",
                color = "gray",
                icon_color = "white"
        )
    
    new_marker = Marker(location = [lat, lon], icon=icon)
    
    new_marker.add_to(sf_map)

sf_map
```
