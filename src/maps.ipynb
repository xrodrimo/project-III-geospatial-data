```python
import folium
from folium import Choropleth, Circle, Marker, Icon, Map
from folium.plugins import HeatMap, MarkerCluster
import pandas as pd
```

```python
latsf = 37.77051
lonrw = -122.401581
```

```python
dfsf = pd.read_csv("../data/coord_SF.csv")
dfrw = pd.read_csv("../data/coord_RW.csv")
dfcu = pd.read_csv("../data/coord_cu.csv")
dfny = pd.read_csv("../data/coord_bony.csv")
```

### Circle San Francicsco

```python
# 1. Map
latsf = 37.77051
lonsf = -122.401581
sf_map = Map(location = [latsf, lonsf], zoom_start = 15)
folium.Circle(
    location=[latsf, lonsf],
    radius=2500,
    color="black",
    weight=1,
    fill_opacity=0.3,
    opacity=1,
    fill_color="red",
    fill=False,  # gets overridden by fill_color
    tooltip="I am in 2.000m",
).add_to(sf_map)
folium.Circle(
    location=[latsf, lonsf],
    radius=1000,
    color="black",
    weight=1,
    fill_opacity=0.3,
    opacity=1,
    fill_color="blue",
    fill=False,  # gets overridden by fill_color
    tooltip="I am in 1000m",
).add_to(sf_map)
folium.Circle(
    location=[latsf, lonsf],
    radius=500,
    color="black",
    weight=1,
    fill_opacity=0.3,
    opacity=1,
    fill_color="green",
    fill=False,  # gets overridden by fill_color
    tooltip="I am in 500m",
).add_to(sf_map)

# 1. Loop the dataframe
    # if the day: holiday
        # icon diferent
    # marker
    # marker.add_to(map)
    
dfsf.sample()

# How to loop a df?
for index, row in dfsf.iterrows():
    
    # 1. location of the accident
    lat = row["lat"]
    lon = row["lon"]
    
    #dict_coordinates = {"location": [row["lat"], row["lon"]]}
        # Marker(location = [])
    
    if row["Type"] == "Standard Education":
        icon = Icon (
            icon = "map-pin",
            prefix = "fa",
            color = "orange",
            icon_color = "black"
        )
        
    elif row["Type"] == "Private Education":
        icon = Icon (
                icon = "map-pin",
                prefix = "fa",
                color = "green",
                icon_color = "white"
        )

    elif row["Type"] == "Night club":
        icon = Icon (
                icon = "globe",
                prefix = "fa",
                color = "white",
                icon_color = "purple"
        ) 
            
    elif row["Type"] == "Starbucks":
        icon = Icon (
                icon = "coffee",
                prefix = "fa",
                color = "green",
                icon_color = "white"
        )
                
    elif row["Type"] == "Design Companies":
        icon = Icon (
                icon = "paint-brush",
                prefix = "fa",
                color = "white",
                icon_color = "dark-blue"
        ) 
                   
    elif row["Type"] == "Vegan":
        icon = Icon (
                icon = "refresh",
                prefix = "fa",
                color = "green",
                icon_color = "blue"
        )   
                 
    elif row["Type"] == "Stadium":
        icon = Icon (
                icon = "futbol",
                prefix = "fa",
                color = "darkgreen",
                icon_color = "white"
        )   
                 
    elif row["Type"] == "Pet Grooming Service":
        icon = Icon (
                icon = "sign-language",
                prefix = "fa",
                color = "blue",
                icon_color = "black"
        )
        
    new_marker = Marker(location = [lat, lon], icon=icon)
    
    new_marker.add_to(sf_map)
sf_map
```

```python
sf_map.save("sf_map.html")
```
