```python
import requests
import json
import pandas as pd
import os
from dotenv import load_dotenv
load_dotenv()
token = os.getenv("token")
```

https://location.foursquare.com/developer/reference/place-search

```python
def requests_for_foursquare (query, lat, lon, radius=500, limit=5):
    url = f"https://api.foursquare.com/v3/places/search?query={query}&ll={lat}%2C{lon}&radius={radius}&sort=RELEVANCE&limit={limit}"
    headers = {
        "accept": "application/json",
        "Authorization": token
    }
    try:
        return requests.get(url, headers=headers).json()
    except:
        print("no :(")
        
def extracting_from_one_element (response):
    name = response["name"]
    distance = response["distance"]
    lat = response["geocodes"]["main"]["latitude"]
    lon = response["geocodes"]["main"]["longitude"]
    
    small_dict = {
    "name": name,
    "distance": distance,
    "lat": lat,
    "lon": lon
}
    return small_dict
```

```python
# Middle school nearby
lat = 37.77051
lon = -122.401581
query = "middle school"
response = requests_for_foursquare(query, lat, lon, radius=2000, limit=5)

def transform_request_into_df (response):
    list_of_dictionaries = [extracting_from_one_element (element) for element in response["results"]]
    return pd.DataFrame(list_of_dictionaries)
list_of_dictionaries = [extracting_from_one_element (element) for element in response["results"]]
df_mch = pd.DataFrame(list_of_dictionaries)
value_to_add = 'Middle school'  # Replace 'Same Value' with the value you want to add
df_mch['Type'] = value_to_add
df_mch
```
